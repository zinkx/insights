```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

## sigmoid 

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
plt.plot(x,sigmoid(x),color = 'teal')
plt.savefig('sigmoid.png')
```

## tanh

```python
x = np.linspace(-10,10,1000)
```

```python
plt.plot(x,np.tanh(x),label = '$tanh(x)$', color = 'teal')
plt.plot(x,2*sigmoid(x)-1, '--', label = '$2*sigmoid(x)-1$',color = 'goldenrod')
plt.legend()
plt.savefig('tanh.png')
```

## ReLU

```python
def ReLU(x):
    a = np.copy(x)
    a[a<0]=0
    return a
```

```python
plt.plot(x,ReLU(x),color = 'teal')
plt.savefig('ReLU.png')
```

## leaky ReLU

```python
alpha = 0.1
```

```python
def leaky_ReLU(x, alpha = 0.1):
    return np.maximum(alpha*x,x)
```

```python
plt.plot(x,ReLU(x),'--',color= 'goldenrod')
plt.plot(x,leaky_ReLU(x), color = 'teal')
plt.savefig('leaky_ReLU.png')
```
