```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,100)
```

## sigmoid 

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def first_order_derivative_sigmoid(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
plt.plot(x,sigmoid(x),color = 'teal',label = '$sigmoid(x)$')
plt.plot(x,first_order_derivative_sigmoid(x),color = 'cadetblue', label = 'gradient $sigmoid(x)$', linestyle ='dotted')
plt.legend()
plt.savefig('sigmoid.png') 
```

## tanh

```python
def first_order_derivative_tanh(x):
    return 1- np.tanh(x)*np.tanh(x)
```

```python
plt.plot(x,np.tanh(x),label = '$tanh(x)$', color = 'teal')
plt.plot(x,2*sigmoid(x)-1, '--', label = '$2*sigmoid(x)-1$',color = 'goldenrod')
plt.plot(x,first_order_derivative_tanh(x),label = 'gradient $tanh(x)$', color = 'cadetblue',linestyle = 'dotted')

plt.legend()

plt.savefig('tanh.png') 
```

## ReLU

```python
def ReLU(x):
    a = np.copy(x)
    a[a<0]=0
    return a
```

```python
def first_order_derivative_ReLU(x):
    a = np.ones_like(x)
    a[x<=0]= 0
    return a
```

```python
plt.plot(x,ReLU(x),color = 'teal',label = '$ReLU(x)$')
# we don't want the plot of the derivative to be connected, hence the two lines for x<=0 and x>0 
plt.plot(x[x<=0],first_order_derivative_ReLU(x[x<=0]),color = 'cadetblue', linestyle = 'dotted')
plt.plot(x[x>0],first_order_derivative_ReLU(x[x>0]),color = 'cadetblue', linestyle = 'dotted',label = 'gradient $ReLU(x)$')
plt.legend()
plt.savefig('ReLU.png')
```

## leaky ReLU

```python
alpha = 0.1
```

```python
def leaky_ReLU(x, alpha = 0.1):
    return np.maximum(alpha*x,x)
```

```python
def first_order_derivative_leaky_ReLU(x, alpha = 0.1):
    a = np.ones_like(x)
    a[x<=0]= alpha
    return a
```

```python
plt.plot(x,ReLU(x),'--',color= 'goldenrod', label = '$ReLU(x)$')
plt.plot(x,leaky_ReLU(x), color = 'teal', label = '$leaky ReLU(x)$')
plt.plot(x[x<=0],first_order_derivative_leaky_ReLU(x[x<=0]), color = 'cadetblue', label = 'gradient $leaky ReLU(x)$',linestyle = 'dotted')
plt.plot(x[x>0],first_order_derivative_leaky_ReLU(x[x>0]), color = 'cadetblue',linestyle = 'dotted')
plt.legend()
plt.savefig('leaky_ReLU.png')
```
